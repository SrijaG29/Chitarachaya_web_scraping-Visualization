Importing pandas and numpy for data manipulation.

```python
import pandas as pd
import numpy as np
```

Reading csv file and converting it into dataframe.

```python
df = pd.read_csv('/content/items.csv',index_col=0)
df = df.reset_index()
df.shape
```

```python
# df.dtypes
df.head(30)
```

**Data Cleaning.**

We will handle category column by using map.

```python
def category_handle(x):
  return x[1:]
y = df['category'].map(category_handle)
df['category'] = y
df
```

```python
df.groupby('delivery_time').delivery_time.count()
```

Converting all (None,) values in delivery_time column into mode of delivery_time column.

```python
def delivery_time_handle(x):
  if x == '(None,)':
    return 'Delivered between 1-3 days for Delhi/NCR or 3-4 days for rest of India\n'
  else:
    return x
y = df['delivery_time'].map(delivery_time_handle)
df['delivery_time'] = y
df
```

We will divide delivery_time column into 2 different columns.

1.delivery_Delhi/NCR

2.delivery_rest_India

```python
def delivery_time_handle(x):
  df['delivery_Delhi/NCR'] = x['delivery_time'][18:21]
  df['delivery_rest_India'] = x['delivery_time'][44:47]
  return x
y = df.apply(delivery_time_handle,axis=1)
df
```

Dropiing delivery_time column.

```python
df.drop(columns= 'delivery_time',inplace = True)
```

```python
df
```

```python
df.groupby('no_of_reviews').no_of_reviews.count()
```

We are replace None and no reviews values with '0'

```python
def no_of_reviews_handle(x):
  # print(x.strip())
  x = x.strip()
  if x == '(None,)' or x == 'No reviews':
    return '0'
  else:
    return x[0]
y = df['no_of_reviews'].map(no_of_reviews_handle)
df['no_of_reviews'] = y
df
```

```python
df.groupby('no_of_reviews').no_of_reviews.count()
```

We are price values from range values to max value.

```python
def price_handle(x):
  x = x.split('–')
  return x[len(x) -1][4:]
y = df['price'].map(price_handle)
df['price'] = y
df
```

we are removing commas in price column.

```python
df['price'] = df['price'].str.replace(',','')
df
```

we are removing 'e)' in price column and replacing it will '0.00'.

```python
def price_e_handle(x):
  if 'e)' in x:
    return '0.00'
  return x.strip()
df['price'] = df['price'].map(price_e_handle)
df
```

```python
df.groupby('rating').rating.count()
```

We are replacing None value in raiting with '0'.

```python
def rating_handle(x):
  if x == '(None,)':
    return '0'
  else:
    return x[-10:-6]
y = df['rating'].map(rating_handle)
df['rating'] = y
df
```

In order to perform heatmap we need to convert all the columns values into numbers.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder={}
heat_map = df.copy()
heat_map
```

```python
heat_map.dtypes
```

We are performing heatmap only on four columns:

-> category: dtype['object']

-> no_of_reviews : dtype['object']

-> price	: dtype['float64']

->rating : dtype['float64']

as category and no_of_reviews have object datatype we will performe label encoding on these to columns.

```python
label_encoder['category'] = LabelEncoder()
heat_map['category'] = label_encoder['category'].fit_transform(heat_map['category'])
label_encoder['no_of_reviews'] = LabelEncoder()
heat_map['no_of_reviews'] = label_encoder['no_of_reviews'].fit_transform(heat_map['no_of_reviews'])
heat_map
```

```python
selected_columns = heat_map[['category', 'no_of_reviews', 'price', 'rating']]
correlation_matrix = selected_columns.corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='PuBu')
plt.title('Heatmap of Selected Columns')
plt.show()
```

Conclusion:

-> no_of_reviews and rating have to most dependency on each other.
